# Validación del Modelo en Datos Futuros (Simulación Temporal)

Este notebook valida el modelo ganador (XGBoost) con datos posteriores a la fecha de corte de entrenamiento.

## Simulación Temporal:

**Contexto:** Simulamos que es el 20 de septiembre de 2025 y queremos evaluar cómo se comportaría el modelo con datos "futuros" reales.

- **Fecha de entrenamiento:** Hasta 2025-09-20 23:59:59
- **Fecha de validación:** Desde 2025-09-21 00:00:00 en adelante
- **Tipo de validación:** Out-of-sample temporal (datos completamente no vistos)

## Objetivos:

1. **Validar rendimiento real** del modelo en datos no vistos temporalmente
2. **Detectar degradación** (model drift) comparando con métricas de test
3. **Evaluar robustez** ante condiciones futuras
4. **Comparar** predicciones con demanda real observada
5. **Decidir** si es necesario reentrenar el modelo

## Periodo de Validación:
- **Fecha inicio validación**: 2025-09-21 00:00:00
- **Datos disponibles**: Según archivo `features_validation.parquet`
- **Dataset**: Generado por notebook 03_exploratory_analysis.ipynb

```python
# Importar bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import json
import joblib
import warnings
from datetime import datetime

warnings.filterwarnings('ignore')

# Configuración de visualización
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
plt.style.use('default')
sns.set_theme(style="whitegrid")

# Sklearn
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error

print("Bibliotecas cargadas")
print(f"Fecha de simulación: 2025-09-20 (punto de corte de entrenamiento)")
print(f"Fecha actual del sistema: {datetime.now().strftime('%Y-%m-%d')}")
```

## Cargar Datos de Validación

```python
# Cargar dataset de validación (datos desde 2025-09-21 en adelante)
# Este archivo fue creado por el notebook 03_exploratory_analysis.ipynb
df_validation = pd.read_parquet('artifacts/data/validation_models/features_validation.parquet')
df_validation['datetime'] = pd.to_datetime(df_validation['datetime'])

print("="*80)
print("DATOS DE VALIDACIÓN (SIMULACIÓN: Prediciendo el 'futuro' desde 2025-09-21)")
print("="*80)
print(f"\nRegistros de validación: {len(df_validation):,}")
print(f"Rango de fechas: {df_validation['datetime'].min().date()} a {df_validation['datetime'].max().date()}")
print(f"Días de validación: {(df_validation['datetime'].max() - df_validation['datetime'].min()).days}")

if len(df_validation) == 0:
    print("\nERROR: No hay datos de validación disponibles")
    print("   Ejecuta el notebook 03_exploratory_analysis.ipynb primero para generar features_validation.parquet")
else:
    print(f"\nDatos de validación cargados correctamente")
    print(f"  Este dataset contiene datos 'futuros' no vistos durante el entrenamiento")
    print(f"  Fecha de corte de entrenamiento: 2025-09-21 00:00:00")
```

## Cargar Modelo Ganador y Componentes

```python
# Directorios
trained_models_dir = Path('artifacts/trained_models')
analysis_dir = Path('artifacts/analysis')

# Cargar metadata del modelo reentrenado (NB08)
metadata_path = trained_models_dir / 'best_model_final_metadata.json'

if not metadata_path.exists():
    print(f"ERROR: No se encuentra el archivo de metadata del modelo reentrenado")
    print(f"       Ruta esperada: {metadata_path}")
    print(f"\n       Debes ejecutar primero el notebook 08_model_comparison.ipynb")
    print(f"       y completar la sección de reentrenamiento.")
    raise FileNotFoundError(f"Metadata no encontrada: {metadata_path}")

with open(metadata_path, 'r') as f:
    metadata = json.load(f)

modelo_nombre = metadata['modelo']
tipo_modelo = metadata['tipo']
metricas_retrain = metadata['metricas_train_completo']
n_samples_train = metadata['n_samples']
fecha_entrenamiento = metadata['fecha_entrenamiento']

print("="*80)
print("MODELO REENTRENADO A VALIDAR (NB08)")
print("="*80)
print(f"\nModelo: {modelo_nombre}")
print(f"Tipo: {tipo_modelo}")
print(f"Algoritmo: {metadata['algoritmo']}")
print(f"\nEntrenamiento:")
print(f"  Fecha: {fecha_entrenamiento}")
print(f"  Registros: {n_samples_train:,}")
print(f"  Features: {metadata['n_features']}")
print(f"\nMétricas en Train Completo (features_train.parquet completo):")
print(f"  MAE: {metricas_retrain['mae']:.2f} MW")
print(f"  RMSE: {metricas_retrain['rmse']:.2f} MW")
print(f"  MAPE: {metricas_retrain['mape']:.2f}%")
print(f"  R²: {metricas_retrain['r2']:.4f}")

# Cargar pipeline del modelo reentrenado
model_path = trained_models_dir / 'best_model_final.pkl'

if not model_path.exists():
    print(f"\nERROR: No se encuentra el modelo reentrenado")
    print(f"       Ruta esperada: {model_path}")
    print(f"\n       Ejecuta el notebook 08_model_comparison.ipynb")
    print(f"       y completa la sección de reentrenamiento.")
    raise FileNotFoundError(f"Modelo no encontrado: {model_path}")

# Cargar el pipeline
pipeline = joblib.load(model_path)

print(f"\nModelo reentrenado cargado correctamente:")
print(f"  Ruta: {model_path}")
print(f"  El pipeline incluye preprocesamiento y modelo reentrenado con TODOS los datos")
```

## Preparar Datos de Validación

```python
# Preparar datos de validación
# El pipeline espera un DataFrame con las columnas exactas que vio durante entrenamiento

print("\nPreparando datos de validación...")

# Separar features y target
y_val = df_validation['demanda'].values
dates_val = df_validation['datetime'].values

# IMPORTANTE: El pipeline de sklearn requiere que las columnas coincidan exactamente
# con las que vio durante el entrenamiento. Seleccionamos solo esas columnas.
if hasattr(pipeline, 'feature_names_in_'):
    required_features = pipeline.feature_names_in_
    print(f"El modelo requiere {len(required_features)} features específicas")
    
    # Verificar que tenemos todas las columnas necesarias
    available_features = set(df_validation.columns)
    missing_features = set(required_features) - available_features
    
    if missing_features:
        print(f"\nERROR: Faltan {len(missing_features)} columnas requeridas:")
        for feat in sorted(missing_features):
            print(f"  - {feat}")
        raise ValueError(f"Faltan columnas requeridas: {missing_features}")
    
    # Seleccionar solo las columnas que el modelo necesita, en el orden correcto
    X_val = df_validation[required_features]
    print(f"Seleccionadas {len(required_features)} features requeridas")
else:
    # Fallback: usar todas las columnas excepto target y datetime
    X_val = df_validation.drop(columns=['demanda', 'datetime'])
    print(f"ADVERTENCIA: No se pudo determinar features requeridas, usando todas ({len(X_val.columns)})")

print(f"\nDatos de validación preparados:")
print(f"  X_val: {X_val.shape}")
print(f"  y_val: {y_val.shape}")
print(f"  Fecha inicio: {pd.Timestamp(dates_val[0]).date()}")
print(f"  Fecha fin: {pd.Timestamp(dates_val[-1]).date()}")
print(f"\nEl pipeline manejará automáticamente el preprocesamiento")
```

## Realizar Predicciones

```python
print("Realizando predicciones con el pipeline...")

# El pipeline ya maneja todo el preprocesamiento internamente
y_pred = pipeline.predict(X_val)

print(f"\n{len(y_pred):,} predicciones realizadas")
print("El pipeline aplicó automáticamente:")
print("  1. Selección de features")
print("  2. Imputación de valores faltantes")
print("  3. Escalado (si es necesario)")
print("  4. Predicción del modelo")
```

## Evaluación de Rendimiento

```python
# Calcular métricas
mae_val = mean_absolute_error(y_val, y_pred)
rmse_val = np.sqrt(mean_squared_error(y_val, y_pred))
mape_val = mean_absolute_percentage_error(y_val, y_pred) * 100
r2_val = r2_score(y_val, y_pred)

# Métricas del modelo reentrenado (train completo del NB08)
mae_train = metricas_retrain['mae']
rmse_train = metricas_retrain['rmse']
mape_train = metricas_retrain['mape']
r2_train = metricas_retrain['r2']

print("="*80)
print("RESULTADOS DE VALIDACIÓN EN DATOS FUTUROS")
print("="*80)
print("\nContexto:")
print("  - Train Completo: features_train.parquet completo (datos históricos)")
print(f"    → Modelo reentrenado con {n_samples_train:,} registros")
print("  - Validation Set: features_validation.parquet (datos futuros >= 2025-09-21)")
print("\nMÉTRICAS:")
print(f"\n{'Métrica':<15} {'Train Completo':<15} {'Validación':<15} {'Diferencia':<15} {'Estado'}")
print("-"*80)

# MAE
diff_mae = ((mae_val - mae_train) / mae_train) * 100
status_mae = "OK" if abs(diff_mae) < 10 else "ADVERTENCIA" if abs(diff_mae) < 20 else "CRÍTICO"
print(f"{'MAE':<15} {mae_train:>10.2f} MW {mae_val:>10.2f} MW {diff_mae:>10.1f}% {status_mae}")

# RMSE
diff_rmse = ((rmse_val - rmse_train) / rmse_train) * 100
status_rmse = "OK" if abs(diff_rmse) < 10 else "ADVERTENCIA" if abs(diff_rmse) < 20 else "CRÍTICO"
print(f"{'RMSE':<15} {rmse_train:>10.2f} MW {rmse_val:>10.2f} MW {diff_rmse:>10.1f}% {status_rmse}")

# MAPE
diff_mape = ((mape_val - mape_train) / mape_train) * 100
status_mape = "OK" if abs(diff_mape) < 10 else "ADVERTENCIA" if abs(diff_mape) < 20 else "CRÍTICO"
print(f"{'MAPE':<15} {mape_train:>11.2f}% {mape_val:>11.2f}% {diff_mape:>10.1f}% {status_mape}")

# R²
diff_r2 = ((r2_val - r2_train) / r2_train) * 100
status_r2 = "OK" if diff_r2 > -2 else "ADVERTENCIA" if diff_r2 > -5 else "CRÍTICO"
print(f"{'R²':<15} {r2_train:>14.4f} {r2_val:>14.4f} {diff_r2:>10.1f}% {status_r2}")

print("\n" + "="*80)

# Evaluación global
degradation_score = (abs(diff_mae) + abs(diff_rmse) + abs(diff_mape)) / 3

print("\nEVALUACIÓN GLOBAL:")
print(f"\nÍndice de degradación: {degradation_score:.1f}%")

if degradation_score < 5:
    print("\nEXCELENTE: El modelo mantiene su rendimiento en datos futuros")
    print("   Acción: Ninguna. Continuar monitoreando.")
elif degradation_score < 10:
    print("\nBUENO: Degradación mínima detectada")
    print("   Acción: Monitoreo continuo. Considerar reentrenamiento en 1-2 meses.")
elif degradation_score < 20:
    print("\nADVERTENCIA: Degradación moderada detectada")
    print("   Acción: Reentrenar modelo con datos actualizados.")
else:
    print("\nCRÍTICO: Degradación significativa detectada")
    print("   Acción: URGENTE - Reentrenar modelo inmediatamente.")
    print("           Investigar cambios en los datos o distribución.")
```

## Análisis de Errores

```python
# Calcular errores
errors = y_val - y_pred
abs_errors = np.abs(errors)
pct_errors = (errors / y_val) * 100

print("="*80)
print("ANÁLISIS DE ERRORES")
print("="*80)

print("\nEstadísticas de Error:")
print(f"  Error medio: {errors.mean():,.2f} MW")
print(f"  Error mediano: {np.median(errors):,.2f} MW")
print(f"  Desviación estándar: {errors.std():,.2f} MW")
print(f"  Error máximo: {errors.max():,.2f} MW")
print(f"  Error mínimo: {errors.min():,.2f} MW")

print("\nPercentiles de Error Absoluto:")
for p in [50, 75, 90, 95, 99]:
    print(f"  P{p}: {np.percentile(abs_errors, p):,.2f} MW")

print("\nDistribución de Errores:")
bins = [0, 100, 250, 500, 1000, np.inf]
labels = ['<100 MW', '100-250 MW', '250-500 MW', '500-1000 MW', '>1000 MW']
error_dist = pd.cut(abs_errors, bins=bins, labels=labels)
error_counts = error_dist.value_counts().sort_index()

for label, count in error_counts.items():
    pct = (count / len(abs_errors)) * 100
    bar = '█' * int(pct / 2)
    print(f"  {label:<15} {count:>6} ({pct:>5.1f}%) {bar}")

# Detectar outliers
threshold_outlier = 1000 
outliers_idx = abs_errors > threshold_outlier
n_outliers = outliers_idx.sum()

if n_outliers > 0:
    print(f"\nOutliers detectados: {n_outliers} ({(n_outliers/len(abs_errors)*100):.2f}%)")
    print(f"   (Errores > {threshold_outlier} MW)")
else:
    print(f"\nNo se detectaron outliers (error > {threshold_outlier} MW)")
```

## Visualizaciones de Validación

```python
# Crear DataFrame para visualización
df_results = pd.DataFrame({
    'datetime': dates_val,
    'real': y_val,
    'predicho': y_pred,
    'error': errors,
    'error_abs': abs_errors,
    'error_pct': pct_errors
})

df_results['fecha'] = pd.to_datetime(df_results['datetime']).dt.date
df_results['hora'] = pd.to_datetime(df_results['datetime']).dt.hour
df_results['dia_semana'] = pd.to_datetime(df_results['datetime']).dt.day_name()

# Figura con múltiples gráficos
fig, axes = plt.subplots(3, 2, figsize=(18, 14))

# 1. Serie temporal completa (primeros 7 días)
ax = axes[0, 0]
n_samples = min(168, len(df_results))
ax.plot(range(n_samples), df_results['real'].iloc[:n_samples], 
        label='Real', linewidth=2, color='green', alpha=0.8)
ax.plot(range(n_samples), df_results['predicho'].iloc[:n_samples], 
        label='Predicho', linewidth=2, linestyle='--', color='blue', alpha=0.8)
ax.set_title('Primeros 7 Días - Demanda Real vs Predicha', fontsize=12, fontweight='bold')
ax.set_xlabel('Horas')
ax.set_ylabel('Demanda (MW)')
ax.legend()
ax.grid(True, alpha=0.3)

# 2. Scatter plot
ax = axes[0, 1]
ax.scatter(df_results['real'], df_results['predicho'], alpha=0.3, s=10)
min_val = min(df_results['real'].min(), df_results['predicho'].min())
max_val = max(df_results['real'].max(), df_results['predicho'].max())
ax.plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2, label='Perfecto')
ax.set_title(f'Real vs Predicho (R²={r2_val:.4f})', fontsize=12, fontweight='bold')
ax.set_xlabel('Demanda Real (MW)')
ax.set_ylabel('Demanda Predicha (MW)')
ax.legend()
ax.grid(True, alpha=0.3)

# 3. Distribución de errores
ax = axes[1, 0]
ax.hist(errors, bins=50, alpha=0.7, color='red', edgecolor='black')
ax.axvline(0, color='black', linestyle='--', linewidth=2)
ax.axvline(errors.mean(), color='blue', linestyle='--', linewidth=2, label=f'Media: {errors.mean():.1f} MW')
ax.set_title('Distribución de Errores', fontsize=12, fontweight='bold')
ax.set_xlabel('Error (MW)')
ax.set_ylabel('Frecuencia')
ax.legend()
ax.grid(True, alpha=0.3)

# 4. Serie temporal de errores
ax = axes[1, 1]
ax.plot(range(n_samples), df_results['error'].iloc[:n_samples], 
        color='red', linewidth=1.5, alpha=0.7)
ax.axhline(0, color='black', linestyle='--', linewidth=1)
ax.fill_between(range(n_samples), df_results['error'].iloc[:n_samples], 
                alpha=0.3, color='red')
ax.set_title('Serie Temporal de Errores (7 días)', fontsize=12, fontweight='bold')
ax.set_xlabel('Horas')
ax.set_ylabel('Error (MW)')
ax.grid(True, alpha=0.3)

# 5. Error por hora del día
ax = axes[2, 0]
error_by_hour = df_results.groupby('hora')['error_abs'].mean()
ax.bar(error_by_hour.index, error_by_hour.values, alpha=0.7, color='orange', edgecolor='black')
ax.set_title('Error Promedio por Hora del Día', fontsize=12, fontweight='bold')
ax.set_xlabel('Hora del Día')
ax.set_ylabel('Error Absoluto Promedio (MW)')
ax.set_xticks(range(0, 24, 2))
ax.grid(True, alpha=0.3, axis='y')

# 6. Error por día de la semana
ax = axes[2, 1]
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
error_by_day = df_results.groupby('dia_semana')['error_abs'].mean().reindex(day_order)
colors = ['#3498db' if day in ['Saturday', 'Sunday'] else '#2ecc71' for day in error_by_day.index]
ax.bar(range(len(error_by_day)), error_by_day.values, alpha=0.7, color=colors, edgecolor='black')
ax.set_title('Error Promedio por Día de la Semana', fontsize=12, fontweight='bold')
ax.set_xlabel('Día')
ax.set_ylabel('Error Absoluto Promedio (MW)')
ax.set_xticks(range(7))
ax.set_xticklabels(['Lun', 'Mar', 'Mié', 'Jue', 'Vie', 'Sáb', 'Dom'], rotation=45)
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('artifacts/analysis/validation_results.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nGráficos guardados en: artifacts/analysis/validation_results.png")
```

## Análisis Temporal de Degradación

```python
# Error por día
error_daily = df_results.groupby('fecha').agg({
    'error_abs': 'mean',
    'error': 'mean',
    'real': 'mean'
}).reset_index()

error_daily['fecha'] = pd.to_datetime(error_daily['fecha'])

# Visualizar tendencia de error
fig, axes = plt.subplots(2, 1, figsize=(16, 10))

# 1. Error absoluto medio diario
ax = axes[0]
ax.plot(error_daily['fecha'], error_daily['error_abs'], 
        marker='o', linewidth=2, markersize=6, color='#e74c3c')
ax.axhline(mae_train, color='green', linestyle='--', linewidth=2, 
          label=f'MAE Train Completo: {mae_train:.2f} MW')
ax.axhline(mae_val, color='blue', linestyle='--', linewidth=2, 
          label=f'MAE Validación: {mae_val:.2f} MW')
ax.fill_between(error_daily['fecha'], mae_train * 0.9, mae_train * 1.1, 
               alpha=0.2, color='green', label='Rango aceptable (±10%)')
ax.set_title('Evolución del Error Absoluto Medio Diario', fontsize=14, fontweight='bold')
ax.set_xlabel('Fecha')
ax.set_ylabel('MAE Diario (MW)')
ax.legend()
ax.grid(True, alpha=0.3)
ax.tick_params(axis='x', rotation=45)

# 2. Sesgo (bias) diario
ax = axes[1]
colors = ['red' if x < 0 else 'green' for x in error_daily['error']]
ax.bar(error_daily['fecha'], error_daily['error'], color=colors, alpha=0.6, edgecolor='black')
ax.axhline(0, color='black', linestyle='-', linewidth=1)
ax.set_title('Sesgo Diario (Sobrepredicción vs Subpredicción)', fontsize=14, fontweight='bold')
ax.set_xlabel('Fecha')
ax.set_ylabel('Error Medio Diario (MW)')
ax.grid(True, alpha=0.3)
ax.tick_params(axis='x', rotation=45)

# Leyenda personalizada
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='green', alpha=0.6, label='Sobrepredicción (predice más)'),
    Patch(facecolor='red', alpha=0.6, label='Subpredicción (predice menos)')
]
ax.legend(handles=legend_elements)

plt.tight_layout()
plt.savefig('artifacts/analysis/validation_temporal_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nAnálisis temporal guardado en: artifacts/analysis/validation_temporal_analysis.png")

# Estadísticas de sesgo
print("\nANÁLISIS DE SESGO:")
bias = errors.mean()
print(f"\nSesgo global: {bias:,.2f} MW")
if abs(bias) < 50:
    print("Sesgo mínimo - El modelo está bien calibrado")
elif abs(bias) < 100:
    print("Sesgo leve detectado - Monitorear")
else:
    print("Sesgo significativo - Considerar recalibración")

if bias > 0:
    print(f"   Tendencia: SOBREPREDICCIÓN (predice {abs(bias):.2f} MW más de lo real)")
else:
    print(f"   Tendencia: SUBPREDICCIÓN (predice {abs(bias):.2f} MW menos de lo real)")
```

## Guardar Resultados de Validación

```python
# Crear reporte de validación
validation_report = {
    'fecha_validacion': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
    'periodo_validacion': {
        'inicio': str(df_results['datetime'].min()),
        'fin': str(df_results['datetime'].max()),
        'dias': (df_results['datetime'].max() - df_results['datetime'].min()).days,
        'n_registros': len(df_results)
    },
    'modelo': modelo_nombre,
    'tipo': tipo_modelo,
    'algoritmo': metadata['algoritmo'],
    'fecha_entrenamiento': fecha_entrenamiento,
    'n_samples_train': n_samples_train,
    'metricas_train_completo': {
        'mae': float(mae_train),
        'rmse': float(rmse_train),
        'mape': float(mape_train),
        'r2': float(r2_train)
    },
    'metricas_validacion': {
        'mae': float(mae_val),
        'rmse': float(rmse_val),
        'mape': float(mape_val),
        'r2': float(r2_val)
    },
    'degradacion': {
        'mae_diff_pct': float(diff_mae),
        'rmse_diff_pct': float(diff_rmse),
        'mape_diff_pct': float(diff_mape),
        'r2_diff_pct': float(diff_r2),
        'degradation_score': float(degradation_score)
    },
    'analisis_errores': {
        'sesgo': float(bias),
        'error_std': float(errors.std()),
        'error_max': float(errors.max()),
        'error_min': float(errors.min()),
        'outliers': int(n_outliers),
        'outliers_pct': float((n_outliers/len(abs_errors)*100))
    },
    'recomendacion': {
        'estado': 'EXCELENTE' if degradation_score < 5 else 'BUENO' if degradation_score < 10 else 'ADVERTENCIA' if degradation_score < 20 else 'CRÍTICO',
        'accion': 'Ninguna' if degradation_score < 5 else 'Monitoreo continuo' if degradation_score < 10 else 'Reentrenar' if degradation_score < 20 else 'URGENTE - Reentrenar',
        'prioridad': 'Baja' if degradation_score < 10 else 'Media' if degradation_score < 20 else 'Alta'
    }
}

# Comprobar que existe directorio artifacts/validations
validations_dir = Path('artifacts/validations/analysis')
validations_dir.mkdir(parents=True, exist_ok=True)


# Guardar reporte JSON
validation_report_path = validations_dir / 'validation_report.json'
with open(validation_report_path, 'w') as f:
    json.dump(validation_report, f, indent=2)

print(f"\nReporte de validación guardado en: {validation_report_path}")

# Guardar predicciones detalladas
predictions_dir = Path('artifacts/data/validation_models')
predictions_dir.mkdir(parents=True, exist_ok=True)

df_results[['datetime', 'real', 'predicho', 'error', 'error_abs', 'error_pct']].to_parquet(
    predictions_dir / 'validation_predictions.parquet'
)
df_results[['datetime', 'real', 'predicho', 'error', 'error_abs', 'error_pct']].to_csv(
    predictions_dir / 'validation_predictions.csv', index=False
)

print(f"Predicciones guardadas en: {predictions_dir}/validation_predictions.{{parquet,csv}}")

# Resumen por día
daily_summary = df_results.groupby('fecha').agg({
    'real': 'mean',
    'predicho': 'mean',
    'error': 'mean',
    'error_abs': 'mean',
    'error_pct': 'mean'
}).reset_index()
daily_summary.columns = ['fecha', 'demanda_real_media', 'demanda_predicha_media', 
                         'error_medio', 'error_abs_medio', 'error_pct_medio']
daily_summary.to_csv(validations_dir / 'validation_daily_summary.csv', index=False)

print(f"Resumen diario guardado en: {validations_dir}/validation_daily_summary.csv")
```

## Conclusiones y Recomendaciones
