```python
# Listamos el numero de registros de cada tabla
import pandas as pd
import os

carpeta_parquet = "../data_parquet"
list_tabla = ['demanda','generacion','almacenamiento','emision']

for tabla in list_tabla:
    carpeta_tabla = os.path.join(carpeta_parquet, tabla)
    if os.path.exists(carpeta_tabla):
        # Mostramos el count de esa tabla
        df = pd.read_parquet(carpeta_tabla)
        print(f"Tabla {tabla}: {df.shape[0]} registros")
    else:
        print(f"Tabla {tabla}: No se encontró la carpeta")
```
